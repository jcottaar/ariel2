```python
import os
import glob

kgs.sanity_checks_active = True
def fname(k):
    return kgs.temp_dir + 'evaluate_model_'+k+'__'+str(len(train_data))+'.pickle'
for k,v in todo.items():
    print(k)
    print('Old file')
    (inferred_data,model,_) = kgs.dill_load(fname(k))
    kgs.score_metric(inferred_data, train_data)
    print(model.model.starter_model.loaders[0].apply_wavelength_binning.n_mean_pixels)
    print('Remodel from old file')
    print(model.model.model_options.FGS_order)
    print(model.model.model_options.AIRS_order_time)
    print(model.model.model_options.AIRS_order_wavelength)
    model.state = 0
    model._cached_planet_id = None
    files = glob.glob(kgs.loader_cache_dir + '/*')
    for f in files:
        os.remove(f)
    # Precache data
    for d in tqdm(train_data):
        copy.deepcopy(d).load_to_step(5,model.model.starter_model.loaders)
    model.train(train_data)
    inferred_data2 = model.infer(train_data)    
    kgs.score_metric(inferred_data2, train_data)
    print('Remodel from new')
    print(model.model.model_options.FGS_order)
    print(model.model.model_options.AIRS_order_time)
    print(model.model.model_options.AIRS_order_wavelength)
    model = v
    print(model.model.starter_model.loaders[0].apply_wavelength_binning.n_mean_pixels)
    files = glob.glob(kgs.loader_cache_dir + '/*')
    for f in files:
        os.remove(f)
    # Precache data
    for d in tqdm(train_data):
        copy.deepcopy(d).load_to_step(5,model.model.starter_model.loaders)
    model.train(train_data)
    inferred_data3 = model.infer(train_data)    
    kgs.score_metric(inferred_data3, train_data)
```

```python
def basics():
    plt.grid(True)
    plt.box(True)
def plot_two_sensors(func, xl, yl, mod):
    _,ax = plt.subplots(1,2,figsize=(12,6))
    plt.sca(ax[0]);basics()
    plt.title('FGS')
    plt.xlabel(xl)
    plt.ylabel(yl)
    func(slice(0,1), 'blue', 'FGS')
    mod()
    
    plt.sca(ax[1]);basics()
    plt.title('AIRS')
    plt.xlabel(xl)
    func(slice(1,283), 'red', 'AIRS')
    mod()
    
    plt.pause(0.001)
compare1= list(todo.keys())[0]
data1 = inferred_data
for v in [inferred_data3]:    
    train_data_here = train_data     
    data2 = v
    error1 = np.array([d.spectrum - t.spectrum for d,t in zip(data1,train_data_here)])
    error2 = np.array([d.spectrum - t.spectrum for d,t in zip(data2,train_data_here)])
    def plot_prediction_vs_true(slic, col, name):
        x = np.mean(error1[:,slic],1)
        y = np.mean(error2[:,slic],1)
        plt.scatter(x,y,color = col)

    plot_two_sensors(plot_prediction_vs_true, compare1+' (mean)', 'comp'+' (mean)', lambda:plt.axline((0,0),slope=1,color='black'))

    plt.figure(); basics()
    plt.title('AIRS variation per planet')
    plt.xlabel(compare1)
    #plt.ylabel(compare2)
    x=np.sqrt(np.var(error1[:,1:],1))
    y=np.sqrt(np.var(error2[:,1:],1))
    plt.scatter(x, y, color='red')
    plt.axline((0,0), slope=1, color='black')
    plt.pause(0.001)
```
